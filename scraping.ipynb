# Modules

```python
import requests
from bs4 import BeautifulSoup
```

# Scrap the html and turn into a beautiful soup object

```python
# Create a variable with the url
url = 'http://dom.manaus.am.gov.br/'

# Use requests to get the contents
r = requests.get(url)
print (r)

# Get the text of the contents
html_content = r.text
print (html_content)

# Convert the html content into a beautiful soup object
soup = BeautifulSoup(html_content, 'lxml')
```

# Select the website’s title

```python
# View the title tag of the soup object
soup.title
```

# Website title tag’s string

```python
# View the string within the title tag
soup.title.string
```

# First paragraph tag


```python
# view the paragraph tag of the soup
soup.p
```

# The parent of the title tag

```python
soup.title.parent.name
```

# The first link tag

```python
soup.a
```

# Find all the link tags "a"

```python
#soup.find_all('a')
a_tags = soup.find_all('a')
a_tags
#print(len(a_tags))
#first = links[0]
#first.index
```

# Listing all links from a page

```python
links = [link.get('href') for link in a_tags]
links
```

# Getting all links that ends with a patern ".pdf"

```python
#type(links[0])
for i in range(0, len(links)):
    links[i] = str(links[i])
    
pdf_links = []
[pdf_links.append(link) for link in links if link.endswith('.pdf')]
pdf_links
```

# Eliminating duplicate elements

```python
#set type can change the order of the elements
pdf_links = list(set(pdf_links))
pdf_links
#print(len(unicos))
```
